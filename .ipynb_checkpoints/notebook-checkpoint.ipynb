### Load Packages


```python
from datetime import datetime
import json
import requests
import pandas as pd
import matplotlib.pyplot as plt

```

# Gathering Data


### Follow example code for pageview and pagecounts link


```python

endpoint_pagecounts = 'https://wikimedia.org/api/rest_v1/metrics/legacy/pagecounts/aggregate/{project}/{access-site}/{granularity}/{start}/{end}'

endpoint_pageviews = 'https://wikimedia.org/api/rest_v1/metrics/pageviews/aggregate/{project}/{access}/{agent}/{granularity}/{start}/{end}'

headers = {
    'User-Agent': 'https://github.com/jameslee0920',
    'From': 'jameslee0920@gmail.com'
}
```

```python
def api_call(endpoint,parameters):
    call = requests.get(endpoint.format(**parameters), headers=headers)
    response = call.json()
    
    return response
```

### Make a loop to go through access types for pageview and page counts


```python
pagecount_access_types = ["all-sites", "desktop-site", "mobile-site"]
pageviews_access_types = ["all-access", "desktop", "mobile-web", "mobile-app"]
monthly_pagecount = []
for access_type in pagecount_access_types:
    pagecount_params = {"project" : "en.wikipedia.org",
                 "access-site" : access_type,
                 "granularity" : "monthly",
                 "start" : "2007120100",
                # for end use 1st day of month following final month of data
                 "end" : "2016080100"
                    }
    monthly_pagecount.extend(api_call(endpoint_pagecounts, pagecount_params)['items'])

monthly_pageview = []
for access_type in pageviews_access_types:
    pageview_params = {"project" : "en.wikipedia.org",
                    "access" : access_type,
                    "agent" : "user",
                    "granularity" : "monthly",
                    "start" : "2015070100",
                    # for end use 1st day of month following final month of data
                    "end" : '2020100100'
                        }
    monthly_pageview.extend(api_call(endpoint_pageviews, pageview_params)['items'])

    
    
```

### Save into JSON FILES


```python
with open('Documents/data-512-a1/pagecounts_200712-201608.json', 'w') as json_file:
    json.dump(monthly_pagecount, json_file)
with open('Documents/data-512-a1/pageviews_201507-202010.json', 'w') as json_file:
    json.dump(monthly_pageview, json_file)
```

# Data Processing

### Convert json data into a data frame format


```python
df_rows = []
for row in monthly_pagecount:
    df_row = pd.DataFrame([{"time":row["timestamp"],
                           "access_type":row["access-site"],
                           "metric_value":row["count"],
                           "metric_type":"pagecount"
                          }])
    df_rows.append(df_row)
for row in monthly_pageview:
    df_row = pd.DataFrame([{"time":row["timestamp"], 
                           "access_type":row["access"],
                           "metric_value":row["views"],
                           "metric_type":"pageview"
                          }])
    df_rows.append(df_row)
df = pd.concat(df_rows).reset_index(drop=True)
    
    
```

### Make month and year from timestamp


```python
df['year'] = df['time'].str[:4]
df['month'] = df['time'].str[4:6]
```

### Change value names to keep consistent between the two api data sources



```python
df.replace({'access_type': {'all-access': 'all_views',
                            'all-sites': 'all_views',
                            'desktop-site': 'desktop_views',
                            'mobile-site': 'mobile_views',
                            'mobile-web': 'mobile_views',
                            'mobile-app': 'mobile_views',
                           }},
           inplace = True
          )

#combine two columns to make one identifying column to be used for pivot later
df['pivot'] = df['metric_type'].str.cat(df['access_type'], sep ="_")
```

### Sum together values of same pivot (mobile app / web to mobile)


```python
df = df.groupby(['year','month','pivot'], as_index=False).agg({'metric_value':'sum'})
```

### Pivot table so that elements in column "pivot" become table names.


```python
pivoted = pd.pivot_table(df, index = ['year','month'], columns = ['pivot'])
pivoted.columns = pivoted.columns.droplevel(0)
pivoted = pivoted.reset_index().rename_axis(None, axis=1)
```

### Remove when desktop views is total (before mobile data)



```python
pivoted.loc[pivoted.pagecount_all_views == pivoted.pagecount_desktop_views, "pagecount_desktop_views"] = pivoted.pagecount_mobile_views
```

### Save dataframe to csv


```python
pivoted.to_csv(path_or_buf="Documents/data-512-a1/en-wikipedia_traffic_200712-202009.csv", index=False)
```

# Start Visualization

### Create a Date column with datetime type value


```python
pivoted['Date'] = pd.to_datetime([f'{y}-{m}-01' for y, m in zip(pivoted.year, pivoted.month)])
```

### Create Graph

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = [20, 10]
plt.figure()
plt.ticklabel_format(style='plain')
plt.plot(pivoted.Date, pivoted.pagecount_all_views/1e6, color = 'black', linestyle = 'dashed', label='total')
plt.plot(pivoted.Date, pivoted.pagecount_desktop_views/1e6, color = 'green', linestyle = 'dashed', label='main site')
plt.plot(pivoted.Date, pivoted.pagecount_mobile_views/1e6, color = 'blue', linestyle = 'dashed', label='mobile site')
plt.plot(pivoted.Date, pivoted.pageview_all_views/1e6, color = 'black', linestyle = 'solid')
plt.plot(pivoted.Date, pivoted.pageview_desktop/1e6, color = 'green', linestyle = 'solid')
plt.plot(pivoted.Date, pivoted.pageview_mobile_views/1e6, color = 'blue', linestyle = 'solid')
plt.legend(loc = "upper left")
plt.title('Page Views on English Wikipedia (x 1,000,000)', fontdict={'fontsize':20})
plt.suptitle('May 2015: a new pageview definition took effect, which eliminated all crawler traffic.  Solid lines mark new definition.', y = 0.08, color = 'darkred')
plt.grid(True)
plt.figure(figsize=(12,8), dpi = 100)



```
